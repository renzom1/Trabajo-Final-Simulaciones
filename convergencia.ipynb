```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
from matplotlib.ticker import MaxNLocator
```

```python
#datos para graficar vs Ecutwfc (sin variar grilla de puntos K)
# eje de las x:
ecut=np.array([20, 40, 60])
#eje de las y (energía)
ecutsalida=np.array([-2275.70234026, -2343.10598310, -2343.22010502])


#datos para graficar vs Kpoints (con Ecutwfc fijo en algún valor)
# eje de las x:
kpoints=np.array([1,2,3,4,5])
#ksalida=np.array([-100, -40,-60,-50])
#ksalida=np.array([-902.21217164 , -902.2584303,-902.26564037,-902.26637991,-902.26688258])
ksalida=np.array([ -2343.24476884 , -2343.51288177 ,-2343.51594146, -2343.51600580,-2343.51600933])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 0.8, 1])

plt.plot(ecut,ecutsalida,color="k",marker="o",linestyle="dashed", label="grilla 1x1x1")
ax1.set_xlabel("ecutwfc (Ry)")
ax1.set_ylabel("E (Ry)")
#ax1.set_ylim(([-902.4,-902]))
#ax1.set_xlim(([40,190]))
ax1.set_title("E vs ecutwfc")

plt.legend(fontsize=12,edgecolor="w"),

ax2 = fig.add_axes([1, 0, 0.8, 1])

plt.plot(kpoints,ksalida,color="k",marker="o",linestyle="dashed", label="Ecutwfc=70 Ry")

ax2.set_xlabel("kpoints")
ax2.set_ylabel("E (Ry)")
ax2.set_ylim(([-2343.52,-2343.51]))
#ax1.set_xlim(([40,190]))
ax2.set_title("E vs Kpoints")
plt.legend(fontsize=12,edgecolor="w"),

plt.show()
```
